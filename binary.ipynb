```python
import sys
import os.path
import binascii
import string
import numpy as np
import cv2
import tkinter
from tkinter import filedialog
```

```python
def read_bytes(filename, chunksize=8192):# バイトごとに抽出
    try:
        with open(filename, "r+b") as f:
            while True:
                chunk = f.read(chunksize)
                if chunk:
                    for b in chunk:
                        yield b
                else:
                    break
    except IOError:
        print("ファイル開けないよ")
        sys.exit(0)
```

```python
def main():
    while True:
        root=tkinter.Tk()
        root.attributes('-topmost', True)
        root.withdraw()
        fTyp = [("","*")]
        iDir = os.path.abspath(os.path.dirname("__file__"))
        file = list(filedialog.askopenfilenames(title= "binary→bmp",filetypes = fTyp,initialdir = iDir))
        if file==[]:
            break

        for file_path in file:
            cnt_0,cnt_1,cnt,cnt_line,list_cnt,num_cnt=1,0,0,0,0,0
            list_instance=[]
            list_num=[]

            for byte in read_bytes(file_path):# 読み込んだファイルごと
                
                if cnt_0==1:#最初の1バイト
                    list_cnt+=1
                    #print(list_cnt,"文字目")
                    cnt_0=0
                    
                elif cnt_1==1:#5バイト目
                    #print(format(byte,'b').zfill(8))
                    list_instance[len(list_instance):len(list_instance)] = list(str(format(byte,'b').zfill(8)))
                    cnt_1=0
                    cnt_line+=1
                    if cnt_line%40==0:
                        list_instance = [int(s) for s in list_instance]
                        array_instance=np.array(list_instance)
                        array_instance=np.reshape(array_instance,(-1,40))
                        array_instance=array_instance*255
                        list_num.append(array_instance)
                        list_instance=[]
                        #print("\n")
                        cnt_0=1
                    
                else :#1から4バイト
                    cnt+=1
                    #print(format(byte,'b').zfill(8),end="")
                    list_instance[len(list_instance):len(list_instance)] = list(str(format(byte,'b').zfill(8)))
                    if cnt==4:
                        cnt=0
                        cnt_1=1
                
            if list_num is not None:
                if not os.path.isdir("./data"):
                    os.mkdir("./data")
                dirnum=len(os.listdir("./data/"))
                #dirnum= "{0:02d}".format(dirnum)
                os.mkdir("./data/"+str(dirnum))
                for num in list_num:
                    num_cnt+=1
                    zero_i = "{0:05d}".format(num_cnt)
                    filepath="./data/"+str(dirnum)+"/"+zero_i+".bmp"
                    cv2.imwrite(filepath,num)
                    print("完了:",filepath)
```

```python
main()
```
